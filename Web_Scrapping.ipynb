**Web Scrapping**

```python
# from selenium import webdriver
from   bs4 import BeautifulSoup
import pandas as pd
```

```python
import urllib
import urllib.request
```

```python
from urllib.request import Request, urlopen
headers = {'User-Agent': 'Mozilla/5.0'}
```

```python
web='https://www.allrecipes.com/recipes-a-z-6735880'
```

```python
request = Request(web, headers=headers)
 # url = urllib.request.urlopen(web, headers={'User-Agent': 'Mozilla/5.0'})
content = urlopen(request).read()
soup = BeautifulSoup(content, 'lxml')
```

```python
linkname=[]
link=[]
recipieLink=[]
```

```python
for a in soup.findAll('li', attrs={'comp link-list__item'}):
  # print(a)
  l=a.find('a',href=True, attrs={'link-list__link type--dog-bold type--dog-link'})
  link.append(l.get('href'))
```

```python
# Scrappig all the recipie links
for web in link:
  request = Request(web, headers=headers)
 # url = urllib.request.urlopen(web, headers={'User-Agent': 'Mozilla/5.0'})
  content = urlopen(request).read()
  soup = BeautifulSoup(content, 'lxml')
  for a in soup.findAll('a',href=True, attrs={'comp mntl-card-list-items mntl-document-card mntl-card card card--no-image'}):
    recipieLink.append(a.get('href'))
  
```

```python
# remove  all the links which has not a recipie 
import re
to_remove = ['gallery','article']

result1 = [i for i in recipieLink if not re.search("|".join(to_remove), i)]
# print(result1)
```

```python
recipie_name=[]
discription=[]
ingredient=[]
recipie=[]
```

```python
nutrion=[]
```

```python

# code to scrap nutrion
for webrec in result1:
  nutri=[]
  # print(webrec)
  request = Request(webrec, headers=headers)
 # url = urllib.request.urlopen(web, headers={'User-Agent': 'Mozilla/5.0'})
  content = urlopen(request).read()
  soup = BeautifulSoup(content, 'lxml')
  for a in soup.findAll('tr', attrs={'mntl-nutrition-facts-summary__table-row'}):

    dictionary = {}
    cal=a.find('td', attrs={'mntl-nutrition-facts-summary__table-cell type--dog'})
    value=a.find('td', attrs={'mntl-nutrition-facts-summary__table-cell type--dog-bold'}) 
    # print(a)
   

    if(cal==None):
      # print(cal.text)
      dictionary['Calories'] = value.text
    else:
      dictionary[cal.text] = value.text
    nutri.append(dictionary)
  nutrion.append(nutri)  
```

```python
import numpy as np
# Save the nutrion in the dataframe
nutrion = np.array(nutrion )
nutrion = pd.DataFrame(nutrion,columns = ['nutrion'])
```

```python

# code to scrap the discription
for webrec in result1:
  ingredien=[]
  recipi=[]
  
  request = Request(webrec, headers=headers)
 # url = urllib.request.urlopen(web, headers={'User-Agent': 'Mozilla/5.0'})
  content = urlopen(request).read()
  soup = BeautifulSoup(content, 'lxml')
  # code to scrap the Recipie name
  h1tag=soup.find('h1',attrs={'comp type--lion article-heading mntl-text-block'})
  recipie_name.append(h1tag.text)
  
  #  #code to scrap the discription name
  ptag=soup.find('p',attrs={'comp type--dog article-subheading'})
  discription.append(ptag.text)
  #Code to scrap discription
  for a in soup.findAll('li', attrs={'mntl-structured-ingredients__list-item '}):
    ingre=a.find('p')
    ingredien.append(ingre.text)
  ingredient.append(ingredien)
  # #code to scrap rcipie steps
  for a in soup.findAll('li', attrs={'comp mntl-sc-block-group--LI mntl-sc-block mntl-sc-block-startgroup'}):
    step=a.find('p')
  # print(step.text)
    recipi.append(step.text)
  recipie.append(recipi)
```

```python
ingredient = np.array(ingredient )
recipie_name = np.array(recipie_name )
discription = np.array(discription )
recipie = np.array(recipie )
```

```python
#Converts all the data to pandas dataframe
ingredient_df = pd.DataFrame(ingredient,columns = ['Ingredient'])
```

```python
import numpy as np
import pandas as pd

#Converts all the data to pandas dataframe
ingredient_df = pd.DataFrame(ingredient,columns = ['Ingredient'])
recipie_df = pd.DataFrame(recipie,columns = ['recipie'])
discription_df = pd.DataFrame(discription,columns = ['discription'])
recipie_name_df = pd.DataFrame(recipie_name,columns = ['recipie_name'])

#concatinate all the data frames
frames = [recipie_name_df, ingredient_df, recipie_df,discription_df,nutrion]

result = pd.concat(frames, axis=1)
result
```

```python

#save the dataframe to csv file for later use 
result.to_csv('RecipieData.csv')
```

**Preprocess** 

the Recipe name colom to check the duplicate recipie

```python
import re
Recipie['recipe_name']=Recipie['recipe_name'].str.lower()
# # Proprocessing the data
# Code to remove the Hashtags from the text
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub(r'\B#\S+','',str(x)))
# Code to remove the links from the text
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub(r"http\S+", "", x))
# Code to remove the Special characters from the text 
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:' '.join(re.findall(r'\w+', x)))
# Code to substitute the multiple spaces with single spaces
# Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub(r'\s+', ' ', x, flags=re.I))
# Code to remove all the single characters in the text
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub(r'\s+[a-zA-Z]\s+', '', x))
# Remove the twitter handlers
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub('@[^\s]+','',x))
Recipie['recipe_name']
```

```python
#Remove all the rows which has duplicate values 

# Using DataFrame.apply() and lambda function 
Recipie = Recipie.apply(lambda x: x.astype(str).str.lower()).drop_duplicates(subset=['recipe_name'], keep='first')
```
